```python
import pandas as pd
```

```python
data = pd.read_csv('us-software-engineer-jobs-zenrows (no NaN).csv')
```

```python
def extract_average_salary(salary_range):
    # Remove non-numeric characters like '$' and ','
    salary_range = salary_range.replace('$', '').replace(',', '').replace('From', '').replace('Up', '').replace('to', '')

    # Check if the range is split correctly
    if ' - ' in salary_range:
        lower, upper = salary_range.split(' - ')
        upper = upper.split()[0]  # Remove "a year" if present
    elif '-' in salary_range:
        lower, upper = salary_range.split('-')
        upper = upper.split()[0]  # Remove "a year" if present
    else:
        # If no range is provided, treat it as a single value
        lower = upper = salary_range.split()[0]  # Remove "a year" if present

    lower = float(lower)
    upper = float(upper)

    # Calculate the average
    average_salary = (lower + upper) / 2
    return average_salary
```

```python
data['average salary'] = data['salary'].apply(extract_average_salary).apply(pd.Series)
```
